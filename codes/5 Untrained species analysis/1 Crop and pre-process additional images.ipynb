```python
from pathlib import Path
import glob, ntpath
from PIL import Image
import numpy as np
import imageio
```

# Set data folder & set masked threshold

```python
folders = glob.glob('Preprocessed/*/')
MASKED_THRESHOLD = 0.05
target_size = (331, 331)
```

```python
folders
```

```python
def get_spc(path):
    return path.split('\\')[1]
get_spc(folders[1])
```

# Crop images using the same protocols as processing test samples

```python
for folder in folders:
    if folder == folders[0]:
        continue
    spc = get_spc(folder)
    print(spc)
    imgs = glob.glob('Preprocessed/{}/*.jpg'.format(spc)) + glob.glob('Preprocessed/{}/*.JPEG'.format(spc))
    
    
    
    for img in imgs:
        im_pil = Image.open(img)
        
        im = np.array(Image.open(img))
        
        w, h = im_pil.size
        Tw, Th = target_size
        M = int(0.48 * w)
        
        if w*0.48 < Tw:
            M = w
        
        tiles = [im[x:x+M,y:y+M] for x in range(0,im.shape[0],M) for y in range(0,im.shape[1],M)]

        count = 0
        for tile in tiles:
            if tile.shape[0] != tile.shape[1]:
                continue
            if (np.sum(np.all(tile == [0, 0, 0], axis=2)) / tile.reshape(-1, 3).shape[0]) > MASKED_THRESHOLD:
                print('skipping by THRESHOLDED', img)
                #continue
            Path('Preprocessed/{}/test_crop/'.format(spc)).mkdir(parents=True, exist_ok=True)                
            imageio.imwrite('Preprocessed/{}/test_crop/{}_crop{}.jpg'.format(spc, ntpath.basename(img).split('.jpg')[0], str(count).zfill(2)), tile)
            count += 1
```
